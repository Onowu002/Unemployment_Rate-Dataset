```python
Unemploy.columns = Unemploy.columns.str.strip()
```

```python
#Group  data by 'Region' and calculate the average unemployment rate for each group
average_unemployment_by_region = Unemploy.groupby('Region')['Estimated Unemployment Rate (%)'].mean()

average_unemployment_by_region
```

```python
# Plotting unemployment rate by region
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y= 'Estimated Unemployment Rate (%)', data=Unemploy)
plt.title('Unemployment Rate by Region in India')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate statistics:

# Calculate additional statistics

# Calculate the median unemployment rate for each region
median_unemployment_by_region = Unemploy.groupby('Region')['Estimated Unemployment Rate (%)'].median()
print("Median Unemployment Rate by Region:\n", median_unemployment_by_region)

# Calculate the standard deviation of the unemployment rate for each region
std_dev_unemployment_by_region = Unemploy.groupby('Region')['Estimated Unemployment Rate (%)'].std()
print("\nStandard Deviation of Unemployment Rate by Region:\n", std_dev_unemployment_by_region)

# Find the region with the highest average unemployment rate
region_with_highest_avg_unemployment = average_unemployment_by_region.idxmax()
print("\nRegion with the highest average unemployment rate:", region_with_highest_avg_unemployment)

# Find the region with the lowest average unemployment rate
region_with_lowest_avg_unemployment = average_unemployment_by_region.idxmin()
print("\nRegion with the lowest average unemployment rate:", region_with_lowest_avg_unemployment)
```

```python
#Group data by month and calculate the average unemployment rate
average_unemployment_by_month = Unemploy.groupby(Unemploy['Date'].dt.month)['Estimated Unemployment Rate (%)'].mean()
average_unemployment_by_month
```

```python
#Extract month for grouping
Unemploy['month'] = Unemploy['Date'].dt.to_period('M')
```

```python
# Plotting average unemployment rate by month
plt.figure(figsize=(10, 6))
plt.plot(average_unemployment_by_month.index, average_unemployment_by_month.values, marker='o', linestyle='-')
plt.xlabel("Month")
plt.ylabel("Average Estimated Unemployment Rate (%)")
plt.title("Average Estimated Unemployment Rate by Month")
plt.xticks(average_unemployment_by_month.index) # Ensure all months are shown on x-axis
plt.grid(axis='y', linestyle='--')
plt.show()
```

#Calculation on Monthly Average Estimated Unemployment Rate, 4th Month is the month with the highest unemployment while 7th Month is the Month with the lowest unemployment rate.
